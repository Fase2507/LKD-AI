```python
df = pd.read_csv('../data/verionislem.csv')
from sklearn.preprocessing import LabelEncoder
```

```python
df.info()
df
```

```python
df["Yaş"] = df["Yaş"].fillna(df["Yaş"].mean())
df["Maaş"] = df["Maaş"].fillna(df["Maaş"].mean())
```

```python
Q1=df["Çalışma Saati"].quantile(0.25)
Q3=df["Çalışma Saati"].quantile(0.75)
iqr=Q3-Q1
down=Q1-1.5*iqr
up=Q3*1.5*iqr
```

```python
(df["Çalışma Saati"]>=down)&(df["Çalışma Saati"]>=up)
print(df)
```

```python
df = df[(df["Çalışma Saati"] >= down) & (df["Çalışma Saati"] <= up)]
label_encoder=LabelEncoder()
df["Şehir"]=label_encoder.fit_transform(df["Şehir"])
df["Departman"]=label_encoder.fit_transform(df["Departman"])
df
```

```python
scaler=MinMaxScaler()
df[["Maaş","Çalışma Saati","Deneyim Yılı"]]=scaler.fit_transform(df[["Maaş","Çalışma Saati","Deneyim Yılı"]])
df
```
